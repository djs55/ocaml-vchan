Vchan is a simple communication protocol over shared memory. It can be used anywhere you can share memory and transmit events. The [ocaml-vchan](https://github.com/mirage/ocaml-vchan) implementation supports communication:

- between 2 Xen domains, in userspace or kernelspace (using Xen grant tables and event channels)
- between 2 Unix processes (using shm and pipes)
- between 2 threads (using shared memory and condition variables)

To begin, link with the `vchan` ocamlfind package and open the module:

````ocaml
open Vchan
````

The protocol logic is expressed as a functor over the means of sharing memory and transmitting events:
```
module Make(A : S.EVENTS)(M: S.MEMORY)(C: S.CONFIGURATION) : S.ENDPOINT with type port = Port.t
```
In the Javascript case we can only use the threads with shared memory and conditional variables version, called `In_memory`:

````ocaml
module V = Vchan.In_memory
````

In vchan, endpoints are identified by the pair of a `domain` (an integer) and a `port` (a string). Not all strings are valid `port`s (because they are used as keys in xenstore), so we must first check our `port` value is valid:

````ocaml
let port = match Vchan.Port.of_string "myport" with
  | `Error _ -> invalid_arg "Vchan.Port.of_string"
  | `Ok x -> x
````

We must start client and server *threads*: note that these threads don't return values (client and server handles) until they have connected to each other:

````ocaml
let server_thread = V.server ~domid:1 ~port ()
````

````ocaml
let client_thread = V.client ~domid:1 ~port ()
````

Now that both client and server threads have been started, they should have connected to each other and the handles should now be available. Note the special use of `lwt` to extract the value from the thread:

````ocaml
lwt server = server_thread
````

````ocaml
lwt client = client_thread
````

Vchan supports the [mirage flow](https://github.com/mirage/mirage/blob/54c8366ab65622d61c402eda1b38c464bc7d7a85/types/V1.mli#L108) interface which operates in terms of raw byte buffers. We need to create one of these and put our string message inside:

````ocaml
let message =
  let txt = "hello" in
  let buf = Cstruct.create (String.length txt) in
  Cstruct.(blit_from_string txt 0 buf 0 (len buf));
  buf
````

We can now write the buffer through the server. Note that we use the `lwt` syntax to block until the write completes. We know it won't block for long because `vchan` uses at least 1 shared page by default with plenty of space for our purposes (in particular it defaults to 1024 bytes for both sending and receiving):

````ocaml
lwt result = V.write server message
````

We can now issue a read through the client. Note again the use of `lwt` to block until the read completes. We know it won't block for long because there are 5 bytes already in the channel:

````ocaml
lwt result = V.read client
````

Hooray, we have read 5 bytes!
